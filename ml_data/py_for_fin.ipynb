```python
"""
One-Dimensional Data Sets
A fundamental and powerful plotting is plt.plot()
In principal it needs two sets of numbers
"""

import numpy as np 
np.random.seed(1000) # Fixes the seed for the random number generator for reproductibility 
y = np.random.standard_normal(20) # Draws the random numbers(y values)
x = np.arange(len(y)) # Fixes the integer (x values)
plt.plot(x, y) # Calls the plot() function with the x and y objects
```

```python
plt.plot(y) 
"""
You can pass NumPy ndarray object to matplotlib functions
Matplotlib is able to interpret the data structures for simplified plotting
Be carefull not to pass a too large and/or complex array
"""
```

```python
"""
Since the majority of thendarray methods return an ndarray object, you can pass the object with a method(even multiple methods, in some cases)


"""
plt.plot(y.cumsum())
```

```python
# One can also set min and max values of each axis 
plt.plot(y.cumsum())
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum()) -1,
         np.max(y.cumsum()) +1)
```

```python
"""
For the sake of readability a plot contains a number of labels(title and labels describing x and y values)
Ploting of discrete data points is done by choosing a diff style option.
The next plot overlays (red) points and a (blue) line width of 1.5 points.
"""

plt.figure(figsize=(10, 6))
plt.plot(y.cumsum(), 'b', lw=1.5) # Plots the data as a line in blue with line width of 1.5 points
plt.plot(y.cumsum(), 'ro') # Plots data as red (thick) dots
plt.xlabel('index')
plt.ylabel('value')
plt.title("A Simple Plot")
```

```python
# Two-Dimensional Data Sets
"""
The following code generates a two-dimensional sample data set as a NumPy ndarray object shape 20 x 2
With standard normally distributed pseudo rand numbers.
"""
y = np.random.standard_normal((20,2)).cumsum(axis=0)

plt.figure(figsize=(10, 6))
plt.plot(y, lw=1.5)
plt.xlabel('index')
plt.ylabel('value')
plt.title("A Simple Plot")
```

```python
"""
plt.legend() accepts diff locality parameters.
0 stands for best location, in the sense that as little data as possible is hidden by the legend
The next plot shows the two data sets, this time with a legend
In the code, the ndarray object is not passed as a whole but the two subsets are accessed separately
Allowing you to attach individual labels to them
"""
plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 1],  lw=1.5, label='2nd')
plt.plot(y, 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title("A Simple Plot")
```

```python
"""
The following code introduces a second y-axis to the plot
The plot now has two different y-axes
The left y-axis is for the first data set while the right y-axis is for the second.
There are also two legends.
"""
fig, ax1 = plt.subplots() # Defines the figure and axis objects
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title("A Simple Plot")
ax2 = ax1.twinx() # Creates a second axis object
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')

# plt.subplots() allows direct access to the underlying plotting object, allowing for one to generate a second plot that shares x-axis.
# The two plots overlay each other.
```

```python
"""
Two separate subplots give even more freedom to handle the two data sets
"""

plt.figure(figsize=(10,6))
plt.subplot(211)
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,0], 'ro')
plt.legend(loc=0)
plt.ylabel('value')
plt.title("A Simple Plot")
plt.subplot(212)
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:,1], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
"""
With the subplot approach onw had the freedom to combine arbitrary kinds of plots that matplotlib offers 
"""
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,0], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel("value")
plt.title("1st Data Set")
plt.subplot(122)
plt.bar(np.arange(len(y)), y[:, 1], width=0.5,
       color='g', label='2nd')
plt.legend(loc=0)
plt.xlabel('index')
plt.title("2nd Data Set")

```

```python
# Static 3D Plotting 
"""
Code artifically generates a plot that resembles a volatility surface.
"""
strike = np.linspace(50, 150, 24)
ttm =  np.linspace(0.5, 2.5, 24)
strike, ttm = np.meshgrid(strike,ttm)
```

```python
strike[:2].round(1)
```

```python
iv = (strike - 100) ** 2 / (100 * strike) / ttm 
iv[:5, :3]
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 6))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2,
                      cmap=plt.cm.coolwarm, linewidth=0.5,
                      antialiased=True)
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
fig.colorbar(surf, shrink=0.5, aspect=5)
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(30, 60)
ax.scatter(strike, ttm, iv, zdir='z', s=25,
          c='b', marker='^')
ax.set_xlabel("strike")
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied valatility')
```

```python
"""
Rolling Statistics 

Rolling Statistics, often called financial indicator or financial studies.
These are basic tools for fianacial chartists and technical traders.
This section works with a single financial time series only.
"""

sym = 'NLSN'
data = pd.DataFrame(data_set[sym].dropna())
data.tail()
```

```python
window = 20

data['min'] = data_set[sym].rolling(window=window).min() # Calculates the rolling minimum values 

data['mean'] = data_set[sym].rolling(window=window).mean() # Calculates the mean value

data['std'] = data_set[sym].rolling(window=window).std() # Calculates the rolling mean value

data['median'] = data_set[sym].rolling(window=window).median() # Calculates the rolling median value 

data['max'] = data_set[sym].rolling(window=window).max() # Calculates the maximum value

data['ewma'] = data_set[sym].ewm(halflife= 0.5, min_periods=window).mean() # Calculates the exponentially weighted moving average.
# With decay in terms of a half life of  0.5
```

```python
data.dropna().head()
```

```python
ax = data[['min','mean', 'max']].iloc[-200:].plot(
     figsize=(10, 6), style=['g--', 'r--', 'g--'], lw=0.8) # Plots three rolling statistics for the final data rows
data_set[sym].iloc[-200:].plot(ax=ax, lw=2.0) # Adds the orriginal time series data to the plot.
```

```python
"""
Technical analysis is using two simple moving averages (SMAs).
The idea is that the trader should go long on a stock when the shorter term SMA is above the longer-term SMA
Doing the opposite also holds true.
"""

```

```python
data['SMA1'] = data[sym].rolling(window=42).mean() # Calculates the values for the short-term SMA

data['SMA2'] = data[sym].rolling(window=252).mean() # Calculates the values for the long-term SMA

data[[sym, "SMA1", "SMA2"]].tail() # Visualizes the stock price data plus the two SMA time series
```

```python
"""
In this content SMAs are only a means to and end. They are used to derive positons to implement a trading strategy.
This visualizes a long position by a value of 1 and a short postion by a value of -1.
Change in a position is triggered by a crossover of the two lines representing the SMA time series 
"""

data.dropna(inplace=True) # Only complete data rows are kept

data['positions'] = np.where(data['SMA1'] > data['SMA2'], # If the shorter-term SMA value is greater than the longer-term...
                            1, # Go long on the stock(put a 1)
                             -1) # Go short on the stock (put-1)

ax = data[[sym, 'SMA1', "SMA2", 'positions']].plot(figsize=(10, 6),
                                                  secondary_y='positions')
ax.get_legend().set_bbox_to_anchor((0.25, 0.85))
```

```python
"""
Correlation Analysis

This is about correlation and not causation. This section shows how to come up with some supporting statistical evidence for.
The S&P and the VIX being (highly negatively correlated)
"""
raw = pd.read_csv('SPY.csv',
                 index_col=0,
                 parse_dates=True)

raw = raw[['Adj Close']].round(3)
raw.index
```

```python
raw_dj = pd.read_csv('VIX.csv',
                 index_col=0,
                 parse_dates=True)

raw_dj = raw_dj[['Adj Close']].round(3)
```

```python
dope = pd.merge(raw, raw_dj,on="Date")
```

```python
dope.rename(columns={'Adj Close_x':"S&P", 'Adj Close_y':"VIX"}, inplace=True)
dope.plot(subplots=True, figsize=(10,6))
```

```python
"""
When plotting (parts of) the time series in a single plot and with adjusted scalings.
The stylized fact of negative correlation between the two indices becomes evident through simple visual inspection.
"""

dope.loc[:'2012-12-31'].plot(secondary_y='VIX', figsize=(10, 6)) # Selects the data until a given value date.
```

```python
"""
Logarithmic Returns 
Statistical analysis relies on returns instead of absolute changes or absolute values.
Lets calculate log returns first before any further analysis takes place.
The next plot shows the high variability of the log returns over time.
For both indices 'volatility-clusters' can be spotted.
In general periods of high volatility can in the stock index are accompanied by the same in the VIX.
"""

retts = np.log(dope / dope.shift(1))

retts.head()
```

```python
retts.dropna(inplace=True)
```

```python
retts.plot(subplots=True, figsize=(10, 6))
```

```python
# pandas scatter_matrix() plotting function comes in handy for visualizations. 

pd.plotting.scatter_matrix(retts,
                          alpha=0.2,
                          diagonal='hist',
                          hist_kwds={'bins':35},
                          figsize=(10,6))
```

```python
"""
OLS Regression
With all these preparations , an ordinary least-squares (OLS) regression analysis is convenient to implement.
The next plot shows a scatter plot of the log returns ans the linear regression line through the cloud of dots.
The slope is obviously negative, proving support for the stylized fact about the negative correlation between the two indices: 
"""

reg = np.polyfit(retts['S&P'], retts['VIX'], deg=1) # This implements a linear OLS regression 
```

```python
ax = retts.plot(kind='scatter', x='S&P', y='VIX', figsize=(10, 6)) # This plots the log returns as a scatter plot

ax.plot(retts['S&P'], np.polyval(reg, retts['S&P']), 'r', lw=2) # To which the linear regression line is added.
```

```python

"""
Correlation
Lets consider correlation measures directly. 
Two such measures are considered:
A static one taking into account the complete data set.
A rolling one showing the correlation for a fixed window over time.
Next plot the correaltion indeed varies over time but that it is always, given parameterization , negative. 
""" 

retts.corr()

ax = retts['S&P'].rolling(window = 252).corr(
                 retts['VIX']).plot(figsize=(10, 6))
ax.axhline(retts.corr().iloc[0, 1],  c='r')
```
