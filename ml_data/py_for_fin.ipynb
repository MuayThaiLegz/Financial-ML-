```python
# Financial Time Series 
import numpy as np 
import pandas as pd
from pylab import mpl, plt
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
file_name = 'sp500_closefull.csv'
```

```python
f = open(file_name, 'r') # Shows the first five rows of the raw data
f.readlines()[:5] # Shows the first five rows of the raw data
```

```python
data = pd.read_csv(file_name, # File name passed to the read_csv function  
                  index_col=0, # Specifies that the first column shall be handled as an index
                  parse_dates=True) # Specifies that the index values are of type datetime
```

```python
data = data.T # Adapting data
```

```python
data_set = data[:20].T # Making the data play nice
```

```python
data_set.plot(figsize=(10, 12), subplots=True) # This visualizes the complete data set via multiple subplots
```

```python
data_set.info() # Give metainformation about the DataFrame object 
```

```python
data_set.describe().round(2) # Provides useful standards statistics per column
```

```python
data_set.aggregate([min,
               np.mean,
               np.std,
               np.median,
               max]).round(2)
```

```python
"""
Changes over Time
Statistical analysis methods are often based on changes over time and not the absolute values themselves. 
There are multiple options to calculate the changes in a series over time.
Including absolute differences, percentage changes and logarithmic (log) returns.
"""
# The absolute differences, for which pandas provodes a special method.

data_set.diff().head() # Provides the absolute changes between two index values 
```

```python
data_set.diff().mean() # Aggregation operations can be applied in addition
```

```python
"""
From a statistics point of view, absolute changes are not optimal because they are dependent on the scale of the time series data. 
Percentage changes are usually preferred.
"""
data_set.pct_change().round(3).head() # pct_change calculates the percentage change between two index values.

```

```python
data_set.pct_change().mean().plot(kind='bar', figsize=(10, 6)) # The mean values of the results are visualized as a bar plot.
```

```python
"""
As an alt to % change, log returns can be used.
Often preferred in a financial context.
""" 
rets = np.log(data_set / data_set.shift(1)) # Calculates the log return in vectorized fashiom
```

```python
rets.head().round(3) # A subset of the results 
```

```python
rets.cumsum().apply(np.exp).plot(figsize=(10, 6)) # Plots the cumulative log returns over time.
# First the cumsum() method is called the np.exp() is applied to the resultssz
```
